```python
#Various functions to process OAI csv data and calculate counts
#-input files: meta_oai.csv dataset with oai university thesis metadata, created by Sara
#-output files: oai_keyword_occurrences_XX.txt, with top words and number of occurrences for nl, en, lang and all keywords

from csv import reader
import operator

wordFreqDict = {}

def getWordStats(word): 
    if(word!=''):
        if word in wordFreqDict:
            currentVal = wordFreqDict[word] + 1
            wordFreqDict.update( { word : currentVal} )
        else:
            wordFreqDict.update( { word : 1 } )

with open('input-data/meta_oai.csv') as f:
    for row in reader(f,delimiter=';'):
        for columnNumber in range(15,55):
            ##en keywords
            getWordStats(row[columnNumber].lower())
        for columnNumber in range(56,114):
            ##lang keywords
            getWordStats(row[columnNumber].lower())
        for columnNumber in range(115,145):
            ##nl keywords
            getWordStats(row[columnNumber].lower())
```

```python
#count occurring words in different columns of dataset

wordFreqSorted = sorted(wordFreqDict.items(), key=operator.itemgetter(1), reverse=True)

file = open("processed-data/oai_keyword_occurrences_ALL.txt","w") 
file.write("itemNumber\tstring\tscore\n")
counter=1
for test in wordFreqSorted:
    print(test[0],test[1])
    file.write(str(counter)+"\t"+str(test[0])+"\t"+str(test[1])+"\n")
    counter+=1
file.close() 
```
