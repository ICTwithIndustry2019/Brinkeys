```python
#various experimental functions to match brinkeys (gold standard) with extracted structure
#-input files: meta_gold.clean.linked-7000.csv (gold standard dataset, linked with almost 7.000 university theses, created by Sara/Alex)
#-input files: brinkies_labels_ppns_hierarchy_v1.tsv (file with ppns<>textual labels, 
#  output of LorentzWS_ProcessBrinkeyStructure.ipynb)
#-input files: brinkeys_gold_and_uni_matched_processed.csv (file with brinkeys, uni ids and unikeys, created by Karen 
#  (original file processed by removing [,],",' and replacing ',' by '|'))

from csv import reader
import operator

wordFreqDict = {}

def getWordStats(word): 
    if(word!=''):
        if word in wordFreqDict:
            currentVal = wordFreqDict[word] + 1
            wordFreqDict.update( { word : currentVal} )
        else:
            wordFreqDict.update( { word : 1 } )

with open('input-data/meta_gold.clean.linked-7000.csv') as f:
    for row in reader(f,delimiter=';'):
        #check if an id (column 30) exists, with a university/thesis ID (i.e., only display linked items)
        if(row[30] != ''):
            for columnNumber in range(11,16):
                ##brinkey keywords
                getWordStats(row[columnNumber].lower())
            
wordFreqSorted = sorted(wordFreqDict.items(), key=operator.itemgetter(1), reverse=True)
```

```python
brinkeyStringDict = {}

with open('processed-data/brinkies_hierarchy_tsv.txt') as f:
    counter = 0
    for row in reader(f,delimiter='\t'):
        #skip first line (column labels)
        if(counter > 0):
            ppn = row[1].lower()
            ppnString = row[0].lower()
            ppnDepth = row[3].lower()
            ppnTree = row[2].lower()
            brinkeyStringDict.update( { ppn : [ppnString,ppnDepth,ppnTree] } )
        counter+=1
```

```python
#file export, with occurrences brinkeys in gold standard set

file = open("processed-data/linked_gold_std_matches_brinkey_v2.tsv","w") 

file.write('ppn\tppn label\toccurrences\tbrinkey depth\n')

for ppnGoldSet in wordFreqSorted:
    for ppnBrinkey in brinkeyStringDict:
        #input file's PPNs sometimes misses initial 0, compare without initial 0
        if(str(ppnGoldSet[0]).lstrip("0")==ppnBrinkey.lstrip("0")):
            file.write(str(ppnBrinkey)+'\t'+str(brinkeyStringDict[ppnBrinkey][0])+'\t'+str(ppnGoldSet[1])+'\t'+str(brinkeyStringDict[ppnBrinkey][1])+'\n')

file.close() 
```

```python
#lookup textual label for a Brinkey PPN

def lookupPpnString(ppn):
    for brPpn in brinkeyStringDict:
        if(brPpn==ppn):
            return brinkeyStringDict[ppn][0]
```

```python
#export for simplified hierarchy visualization (root level, i.e. highest terms in hierarchy, leaf level, i.e. lowest terms in hierarchy)

file = open("processed-data/linked_gold_std_matches_brinkey_for_vizz_simplified_hierarchy.tsv","w") 
file.write('itemNumber\tstring\tscore\tdepth\thierarchy\n')

for ppnGoldSet in wordFreqSorted:
    for ppnBrinkey in brinkeyStringDict:
        if(str(ppnGoldSet[0]).lstrip("0")==ppnBrinkey.lstrip("0")):
            file.write(str(ppnGoldSet[0])+'\t'+str(brinkeyStringDict[ppnBrinkey][0])+'\t'+str(ppnGoldSet[1])+'\t'+str(brinkeyStringDict[ppnBrinkey][1])+'\t'+str(lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[0]))+'\n')
            
file.close() 
```

```python
#export for hierarchy visualization v2 (all hierarchy levels)

file = open("processed-data/linked_gold_std_matches_brinkey_for_vizz_full_hierarchy.tsv","w") 
file.write('itemNumber\tstring\tscore\tdepth\thierarchy_1\thierarchy_2\thierarchy_3\thierarchy_4\thierarchy_5\n')

for ppnGoldSet in wordFreqSorted:
    for ppnBrinkey in brinkeyStringDict:
        if(str(ppnGoldSet[0]).lstrip("0")==ppnBrinkey.lstrip("0")):
            brinkeyHierarchyList = brinkeyStringDict[ppnBrinkey][2].split(',')
            hierarchy1=hierarchy2=hierarchy3=hierarchy4=hierarchy5 = ''
            if(len(brinkeyHierarchyList)>4):
                hierarchy5=lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[4])
            if(len(brinkeyHierarchyList)>3):
                hierarchy4=lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[3])
            if(len(brinkeyHierarchyList)>2):
                hierarchy3=lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[2])
            if(len(brinkeyHierarchyList)>1):
                hierarchy2=lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[1])
            if(len(brinkeyHierarchyList)>0):
                hierarchy1=lookupPpnString(brinkeyStringDict[ppnBrinkey][2].split(',')[0])
            
            file.write(str(ppnGoldSet[0])+'\t'+str(brinkeyStringDict[ppnBrinkey][0])+'\t'+str(ppnGoldSet[1])+'\t'+str(brinkeyStringDict[ppnBrinkey][1])+'\t'+str(hierarchy1)+'\t'+str(hierarchy2)+'\t'+str(hierarchy3)+'\t'+str(hierarchy4)+'\t'+str(hierarchy5)+'\n')
            
file.close()
```

```python
#Find connections between keywords

keywordConnectionDict = {}

#verify: new file!
with open('input-data/Brinkeys_gold_and_uni_matched.tsv') as f:
#with open('/Users/hugo/Github/LorentzWS2018/brinkeys_gold_and_uni_matched_processed.csv') as f:
        counter = 0
        #for row in reader(f,delimiter=','):
        for row in reader(f,delimiter='\t'):
            if(counter > 0):
                keywordStringMatch = False
                for brinkey in row[1].split("|"):
                    for unikey in row[3].split("|"):
                        if(brinkey==unikey):
                            keywordStringMatch = True
                #fields: brinkey,unikey,keywordstringmatch
                keywordConnectionDict.update( { row[1].replace("|",", ") : [row[3].replace("|",", "),keywordStringMatch,row[2].split("-")[0]] } )
            counter += 1

f.close() 
```

```python
#processing: count brinkey and unikey occurrences in matched dataset

brinkeyFreqDict = {}
unikeyFreqDict = {}

def getWordConnectionOccurrences(word,wDict): 
    if(word!=''):
        if word in wDict:
            currentVal = wDict[word] + 1
            wDict.update( { word : currentVal} )
        else:
            wDict.update( { word : 1 } )

def countBrinkeyOccurrences():
    for brinkeyArr in keywordConnectionDict:
        for brinkey in brinkeyArr.split(', '):
            getWordConnectionOccurrences(brinkey,brinkeyFreqDict)

def countUnikeyOccurrences():
    for brinkeyArr in keywordConnectionDict:
        for unikey in keywordConnectionDict[brinkeyArr][0].split(', '):
            getWordConnectionOccurrences(unikey,unikeyFreqDict)
            
countBrinkeyOccurrences();
countUnikeyOccurrences();
```

```python
#export for keyword matching visualization, between brinkey and uni

file = open("processed-data/linked_gold_std_matches_brinkey_for_keyword_similarity.tsv","w") 

file.write('brinkey\tunikey\tkeyword match\tuniversity\tbrinkeyPop\tunikeyPop\n')

def getNumberOfOccurrences(string,arr):
    for keyword in arr:
        if(keyword==string):
            return arr[keyword]
    return 0

def countAllOccurrences(string,arr):
    totalOccurrences = 0
    for singleKey in string.split(', '):
        totalOccurrences += int(getNumberOfOccurrences(singleKey,arr))
    return totalOccurrences
        
for brinkey in keywordConnectionDict:
    brinkeyFreq = countAllOccurrences(brinkey,brinkeyFreqDict)
    unikeyFreq = countAllOccurrences(keywordConnectionDict[brinkey][0],unikeyFreqDict)
    file.write(str(brinkey)+'\t'+str(keywordConnectionDict[brinkey][0])+'\t'+str(keywordConnectionDict[brinkey][1])+'\t'+str(keywordConnectionDict[brinkey][2])+'\t'+str(brinkeyFreq)+'\t'+str(unikeyFreq)+'\n')
    
file.close() 
```
