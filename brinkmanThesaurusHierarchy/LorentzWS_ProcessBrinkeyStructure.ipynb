```python
#functions to map out Brinkey structure and output this to "human readable" and csv files

#-input files: brinkman.20190121.xml: XML version of brinkey structure, created by Erik (KB)
#-output files: brinkies_hierarchy_human_readable.txt: brinkeys in "human readable format"
#-output files: brinkies_hierarchy_csv.txt: PPNs, brinkeys and their structure as CSV data

from lxml import etree
import os

el = etree.parse('input-data/brinkman.20190121.xml')
lemmas = el.xpath("//lemma")

termDict = {}

#find root terms (highest in hierarchy, e.g. "chemie")
def findRootTerms(): 
    #get all lemmas which do NOT have a broader term
    rootTermsArr = el.xpath("/thesaurus/lemmata/lemma[not(broader_term)]")
    counter = 0
    for term in rootTermsArr:
        print("processing term " + str(counter) + " out of " + str(len(rootTermsArr)))
        
        ppn = term.attrib['ppn']
        
        #add base term to dict
        file.write('\nL1. ' + getTermLabel(ppn) + ' (' + ppn +  ')\n')
        fileCsv.write(getTermLabel(ppn)+'\t'+ ppn +'\t' + ppn + '\t'+'1\n')
        
        #quick fix..
        previousString = ppn + ','
        
        #find narrower terms, will be added as lists
        findNarrowerTerm(ppn)
        
        counter+=1      
        
def findNarrowerTerm(ppn):
    narrowerTermArr = el.xpath("/thesaurus/lemmata/lemma[./broader_term[@ppn='"+ppn+"']]")
    
    for term in narrowerTermArr:
        narrowerPpn = term.attrib['ppn']
        
        file.write("  L2. " + getTermLabel(narrowerPpn) + ' (' + narrowerPpn +  ')\n')
        previousString = ppn + ',' + narrowerPpn
        fileCsv.write(getTermLabel(narrowerPpn)+'\t'+ narrowerPpn +'\t'+previousString+'\t'+'2\n')
        
        narrowerTermChildren = []
        
        getNarrowerTerms(narrowerTermChildren,narrowerPpn,3,previousString)

def getNarrowerTerms(termArr,parentTerm,recursionLevel,previousString):
    narrowerTermArr = el.xpath("/thesaurus/lemmata/lemma[./broader_term[@ppn='"+parentTerm+"']]")
    
    if(len(narrowerTermArr) > 0):
        for term in narrowerTermArr:
            
            foundNarrowerTermsArr = []
            narrowerPpn = term.attrib['ppn']
            
            newString = previousString +','+ narrowerPpn
            file.write(printSpacing(recursionLevel) + 'L' + str(recursionLevel) + ". " + getTermLabel(narrowerPpn) + ' (' + narrowerPpn +  ')\n')
            fileCsv.write(getTermLabel(narrowerPpn)+'\t'+narrowerPpn+'\t'+newString+'\t'+str(recursionLevel) + '\n')
            
            getNarrowerTerms(foundNarrowerTermsArr,narrowerPpn,(recursionLevel+1),newString)

def printSpacing(recursionLevel):
    spacing = "  "
    for cnt in range(0,recursionLevel):
        spacing += "  "
    return spacing

def getTermLabel(ppn):
    nameTag = el.xpath("/thesaurus/lemmata/lemma[@ppn='"+ppn+"']/name")
    return nameTag[0].text


#main program    
file = open("processed-data/brinkies_hierarchy_human_readable.txt","w")
fileCsv = open("processed-data/brinkies_hierarchy_tsv.txt","w")

fileCsv.write('brinkey label\tbrinkey ppn\tbrinkey ppn hierarchy\thierarchy depth\n')

findRootTerms()
file.close()
fileCsv.close()
```
